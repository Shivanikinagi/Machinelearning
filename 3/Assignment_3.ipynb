```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv('/content/Admission_Predict_Ver1.1.csv')
subset_data = data[["GRE Score", "TOEFL Score", "CGPA", "Chance of Admit "]].copy()

plt.figure(figsize=(8, 6))
plt.scatter(subset_data["GRE Score"], subset_data["Chance of Admit "], alpha=0.7, color='blue')
plt.title("Scatter Plot: GRE Score vs. Chance of Admit", fontsize=14)
plt.xlabel("GRE Score", fontsize=12)
plt.ylabel("Chance of Admit", fontsize=12)
plt.grid(True)
plt.show()

```

```python
avg_scores = data.groupby("University Rating")[["TOEFL Score", "CGPA"]].mean()
avg_scores.plot(kind="bar", figsize=(10, 6))
plt.title("Bar Plot: Avg TOEFL Score & CGPA by University Rating", fontsize=14)
plt.ylabel("Average Scores", fontsize=12)
plt.xlabel("University Rating", fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis="y")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=subset_data[["GRE Score", "TOEFL Score"]])
plt.title("Box Plot: GRE and TOEFL Score Distribution", fontsize=14)
plt.ylabel("Scores", fontsize=12)
plt.show()
```

```python
research_counts = data["Research"].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(research_counts, labels=["No Research", "Research"], autopct='%1.1f%%', colors=["#ff9999", "#66b3ff"])
plt.title("Pie Chart: Proportion of Research Experience", fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data["Serial No."], data["GRE Score"], label="GRE Score", marker="o")
plt.plot(data["Serial No."], data["TOEFL Score"], label="TOEFL Score", marker="s")
plt.title("Line Chart: GRE and TOEFL Scores Across Serial Numbers", fontsize=14)
plt.xlabel("Serial No.", fontsize=12)
plt.ylabel("Scores", fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```
